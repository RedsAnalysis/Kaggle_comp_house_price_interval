```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
try:
    df_train = pd.read_csv("dataset.csv", index_col="id", parse_dates=["sale_date"])
    df_test = pd.read_csv("test.csv", index_col="id", parse_dates=["sale_date"])
    print("Datasets loaded successfully.")
    print(f"Training data shape: {df_train.shape}")
    print(f"Test data shape: {df_test.shape}")
except FileNotFoundError:
    print("ERROR: Make sure 'dataset.csv' and 'test.csv' are in the same directory as the notebook.")
```

```python
print("\n--- Initial Data Inspection ---")
print("\nFirst 5 rows of the training data:")
display(df_train.head())

print("\nData types and non-null counts:")
df_train.info()

print("\nSummary statistics for numerical columns:")
# Use .T to transpose for easier reading
display(df_train.describe().T)
```

```python
print("\n--- 3. Target Variable Analysis: 'sale_price' ---")
plt.figure(figsize=(16, 6))
```

```python
# Plot 1: Distribution of sale_price
plt.subplot(1, 2, 1)
sns.histplot(df_train['sale_price'], kde=True, bins=50)
plt.title('Distribution of Sale Price (Original)')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
```

```python
# Plot 2: Log-transformed distribution of sale_price
# np.log1p(x) is equivalent to np.log(1+x) - it's better for values close to 0
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df_train['sale_price']), kde=True, bins=50, color='green')
plt.title('Distribution of Sale Price (Log-Transformed)')
plt.xlabel('Log(1 + Sale Price)')
plt.ylabel('Frequency')

plt.suptitle('Sale Price Distribution - Original vs. Log-Transformed', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
print("\n--- 4. Numerical Feature Analysis ---")
numerical_cols = df_train.select_dtypes(include=np.number).columns.tolist()
# Remove sale_price from this list for correlation analysis
numerical_cols.remove('sale_price')

# Correlation heatmap
plt.figure(figsize=(14, 10))
correlation_matrix = df_train[numerical_cols + ['sale_price']].corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.show()
```

```python
# Show top correlations with sale_price
print("\nTop 10 features most correlated with 'sale_price':")
top_corr = correlation_matrix['sale_price'].sort_values(ascending=False).drop('sale_price').head(10)
print(top_corr)
```

```python
# Scatter plots for top correlated features vs. sale_price
print("\nScatter plots for highly correlated features vs. log(sale_price):")
plt.figure(figsize=(18, 10))
for i, col in enumerate(top_corr.index[:4]): # Plot top 4
    plt.subplot(2, 2, i + 1)
    sns.scatterplot(x=df_train[col], y=np.log1p(df_train['sale_price']), alpha=0.3, s=15)
    plt.title(f'{col} vs. Log(Sale Price)')
    plt.xlabel(col)
    plt.ylabel('Log(Sale Price)')
plt.tight_layout()
plt.show()
```

```python
print("\n--- 5. Categorical Feature Analysis ---")

# First, let's confirm the column names
print("Available columns in the DataFrame:")
print(df_train.columns.tolist())
print("-" * 50)

# --- After you run the line above, UPDATE the list below with the correct names ---
# Example: If 'bedrooms' is actually 'beds', change it in the list.
# I'm guessing some common names, but you must verify them.
pseudo_categorical_cols = [
    'bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'condition', 'grade'
]

# A quick check to ensure all columns in our list actually exist in the DataFrame
# This will prevent the KeyError from happening again.
existing_pseudo_categorical = [col for col in pseudo_categorical_cols if col in df_train.columns]
if not existing_pseudo_categorical:
     print("WARNING: None of the specified pseudo-categorical columns were found. Skipping box plots.")
else:
    print(f"\nFound the following pseudo-categorical columns to plot: {existing_pseudo_categorical}")
    
    # Calculate the number of rows needed for the subplot grid
    num_plots = len(existing_pseudo_categorical)
    num_cols = 3 # Let's keep 3 plots per row
    num_rows = (num_plots + num_cols - 1) // num_cols # Ceiling division

    plt.figure(figsize=(20, 5 * num_rows))
    for i, col in enumerate(existing_pseudo_categorical):
        plt.subplot(num_rows, num_cols, i + 1)
        # Check if the column has too many unique values to be plotted reasonably
        if df_train[col].nunique() > 50:
             print(f"Skipping boxplot for '{col}' because it has too many unique values ({df_train[col].nunique()}).")
             plt.text(0.5, 0.5, f"'{col}' has too many unique values\nto plot as a boxplot.", ha='center', va='center')
             plt.gca().set_xticks([])
             plt.gca().set_yticks([])
        else:
            # Order the boxes by median price for better visualization
            order = df_train.groupby(col)['sale_price'].median().sort_values().index
            sns.boxplot(x=col, y=np.log1p(df_train['sale_price']), data=df_train, order=order)
            plt.title(f'Log(Sale Price) by {col}')
            plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

print("\nEDA on categorical features complete.")
```

```python
print("\n--- 6. Date and Geospatial Analysis ---")

# Date analysis: Median sale price over time
# This part likely worked correctly, but we include it for completeness.
try:
    # Ensure sale_month doesn't already exist from a previous run
    if 'sale_month' in df_train.columns:
        df_train = df_train.drop(columns=['sale_month'])
        
    df_train['sale_month'] = df_train['sale_date'].dt.to_period('M')
    monthly_prices = df_train.groupby('sale_month')['sale_price'].median()
    
    # Convert PeriodIndex to DateTimeIndex for plotting
    monthly_prices.index = monthly_prices.index.to_timestamp()

    plt.figure(figsize=(16, 6))
    monthly_prices.plot()
    plt.title('Median House Price Over Time', fontsize=16)
    plt.ylabel('Median Sale Price')
    plt.xlabel('Sale Month')
    plt.show()

except Exception as e:
    print(f"An error occurred during date analysis: {e}")


# --- VERIFY THESE COLUMN NAMES from your df_train.columns.tolist() output ---
LAT_COL = 'latitude'  # <-- Change this if your latitude column is named differently
LON_COL = 'longitude' # <-- Change this if your longitude column is named differently

# Geospatial analysis: lat/long colored by price
if LAT_COL in df_train.columns and LON_COL in df_train.columns:
    print(f"\nPlotting geospatial data using '{LON_COL}' and '{LAT_COL}'...")
    plt.figure(figsize=(12, 12))
    sns.scatterplot(
        x=LON_COL,
        y=LAT_COL,
        data=df_train.sample(20000, random_state=1), # Add random_state for reproducibility
        hue='sale_price',
        palette='viridis',
        alpha=0.6,
        s=20,
        hue_norm=plt.Normalize(vmax=df_train['sale_price'].quantile(0.95)) # Normalize color for better contrast
    )
    plt.title('House Prices by Location', fontsize=16)
    plt.legend(title='Sale Price')
    plt.show()
else:
    print(f"WARNING: Could not find '{LON_COL}' and/or '{LAT_COL}' in the DataFrame columns. Skipping geospatial plot.")
```
